## Experiment 7b-Agglomerative clustering

### Import packages


```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
from sklearn.cluster import AgglomerativeClustering
```

```python
Students_data = pd.read_csv("/content/students_placement_data (1).csv")
Students_data.head()
```

```python
data = Students_data.iloc[:,4:6].values
```

```python
cluster = AgglomerativeClustering(n_clusters= 3, affinity= 'euclidean', linkage= "complete")
cluster.fit_predict(data)
```

### Visualize the clusters

```python
plt.xlabel("inter_Diploma_percentage")
plt.ylabel("B.Tech_percentage")
plt.scatter(data[:,0],data[:,1], c=cluster.labels_, cmap='rainbow')
```

###Using dendrograms to visualize the data.

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(15,7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data,method='ward'))
```
